```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\preet\Desktop\housing.csv")
df
# Histogram with KDE
sns.histplot(data=df, x="housing_median_age", kde=True, bins=30)
plt.title("Distribution of Housing Prices")
plt.show()
```

```python
from scipy.stats import skew, kurtosis
housing_median_age_skew = skew(df["housing_median_age"])
housing_median_age_kurt = kurtosis(df["housing_median_age"])
print("Skewness:",housing_median_age_skew)
print("Kurtosis:",housing_median_age_kurt)
```

```python
sns.countplot(data=df, x="housing_median_age")
plt.title("Count housing_median_age")
plt.xticks(rotation=45)
plt.show()
```
